### Appy SVD method

```python
plt.style.use('classic')
image = Image.open('/content/drive/MyDrive/cursos_platzi/Linear-algebra-in-python/images/image2.png')
plt.imshow(image)
```

```python
image = image.convert('LA') #gray scale
matrix_image = np.array(list(image.getdata(band=0)), float)
matrix_image.shape = (image.size[1], image.size[0])
```

```python
#Decomposition

U, D, V = np.linalg.svd(matrix_image)
```

```python
print(matrix_image.shape)
print(U.shape)
print(D.shape)
print(V.shape)
```

Many singular values don't represent important information. But we do need some of them, let's see what happens if we take a few s.v's:

```python
image_reconstruction = np.matrix(U[:, :1])*np.diag(D[:1])*np.matrix(V[:1,:])
plt.imshow(image_reconstruction, cmap='gray')
plt.show()
```

### Determine how many singular values are needed

```python
#for i in range(len(D)):
for i in range(150):
  image_reconstruction = np.matrix(U[:, :i])*np.diag(D[:i])*np.matrix(V[:i,:])
  plt.imshow(image_reconstruction, cmap='gray')
  title = '#Singular values %s' %i
  plt.title(title)
  plt.show()
```
